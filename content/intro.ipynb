# Modellbildung und Simulation

Dies ist ein interaktives Übungsbuch zur Vorlesung __Modellbildung und Simulation__ an der [Hochschule Bonn-Rhein-Sieg](https://www.h-brs.de). Es dient zur Übermittlung von Kenntnissen der Modellierung und Simulation ingenieurstechnischer Probleme unter der Verwendung von Matlab bzw. Octave.

Es ist ein gemeinsames Projekt von [Jan Kleinert](https://www.dlr.de/sc/desktopdefault.aspx/tabid-1192/1635_read-34880/sortby-lastname/), [Dirk Reith](https://www.h-brs.de/de/emt/dirk-reith) und [Gerd Steinebach](https://www.h-brs.de/de/emt/prof-dr-gerd-steinebach), das sich gerade im Aufbau befindet. Wir freuen uns über jegliche Art von Feedback!

````octave
x = linspace(-8, 8, 41)';
y = x;
[X, Y] = meshgrid(x, y);
R = sqrt(X.^2 + Y.^2) + eps;
Z = sin(R)./ R;
surf(X, Y, Z);
````

Im Gegensatz zu Matlab-Liveskripten wird hier auf die kostenfreie Matlab-Alternative [Octave](https://www.gnu.org/software/octave/) zusammen mit dem [Project Jupyter](https://jupyter.org/) gebaut, um mit frei verfügbaren Software-Tools frei verfügbare Lerninhalte bereitzustellen.
